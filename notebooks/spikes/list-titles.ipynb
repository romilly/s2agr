```python
import csv

from s2agr.persistence.database_catalogue import DatabaseCatalogue, production_connection
```

```python
connection = production_connection()
catalogue = DatabaseCatalogue(production_connection())
```

```python
with open('papers.txt','w') as pt:
    for _, title in catalogue.query("select paper_id, (s2ag_json_text->>'abstract') from paper"):
        if title is not None:
            pt.write(title)
            pt.write('\n')
```

```python
import sys
from collections import defaultdict


def count_tokens(filename):
    result = defaultdict(int)
    with open(filename) as tf:

        for title in tf:
            tokens = title.split()
            for token in tokens:
                result[token] += 1
    return result

if __name__ == '__main__':
    print(count_tokens(sys.argv[1]))
```

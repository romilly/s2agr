```python
import json
import psycopg2
from datetime import datetime

# Connect to your postgres DB
conn = psycopg2.connect("dbname='arxiv' user='pi' host='s2ag' password='simple'")
```

```python

# Open a cursor to perform database operations
cur = conn.cursor()
```

```python

# Create a table with an id, categories, creation timestamp, and json field
cur.execute("""
    CREATE TABLE IF NOT EXISTS json_records (
        id TEXT PRIMARY KEY,
        categories TEXT,
        created_at TIMESTAMP,
        record JSONB
    )
""")

# Open the file and load each record into the database
with open('../data/downloads/arxiv-metadata-oai-snapshot.json', 'r') as f:
    count = 0
    for line in f:
        record = json.loads(line)
        id = record.get('id')
        categories = record.get('categories')
        created_at = datetime.now()
        cur.execute("""
            INSERT INTO json_records (id, categories, created_at, record)
            VALUES (%s, %s, %s, %s) ON CONFLICT DO NOTHING
        """, (id, categories, created_at, json.dumps(record)))
        count += 1
        if 0 == count % 1000:
            conn.commit()
            print(count, id)

# Commit the transaction
conn.commit()

# Close the connection
cur.close()
conn.close()
```

```python
import time
from s2ag.librarian import Librarian, WebResearcher
from s2ag.persistence.database_catalogue import DatabaseCatalogue, test_connection
from s2ag.requester import WebRequester
```

```python
requester = WebRequester()
catalogue = DatabaseCatalogue(test_connection())
librarian = Librarian(WebResearcher(requester), catalogue)
fly_paper_id = 'e2e1aa8b8b1dfed9c65589d5293acbae4cbe061a' # small set of citations
paper = librarian.get_paper(fly_paper_id)
```

```python
citations = paper.citations
```

```python
def iterate(paper_id):
        paper = librarian.get_paper(paper_id)
        citations = paper.citations
        for citation in citations:
                next_paper = citation['paperId']
                if next_paper is None:
                        print('Oops!', paper_id, len(citations), citation)
                        continue
                if catalogue.knows(next_paper):
                        continue
                print(next_paper)
                time.sleep(3.1)
                librarian.get_paper(next_paper)
```

```python
for paper_id in catalogue.paper_ids():
        iterate(paper_id)
print('done!')
```
